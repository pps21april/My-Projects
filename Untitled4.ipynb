```python
def merge(nums1, m, nums2, n) :
        """
        Do not return anything, modify nums1 in-place instead.
        """
        new_nums = [0]*(m+n)

        if m==0 :
            return nums2[:]
        if n==0:
            return nums1[:]
        if nums1[m-1]<=nums2[0]:
            return nums1[:] + nums2[:]

        i=0
        j=0
        k=0

        while i < m and j < n:
             if nums1[i]>nums2[j]:
                new_nums[k]=nums2[j]
                k+=1
                j+=1
             else:
                new_nums[k]=nums1[i]
                k+=1
                i+=1

        while i<m:
          new_nums[k] = nums1[i]
          i+=1
          k+=1

        while j<n:
          new_nums[k] =nums2[j]
          j+=1
          k+=1

        return new_nums


merge([0],1,[0],1)
```
