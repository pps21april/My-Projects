##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  


Outliers are values which fall outside the range = (mean - 3(standard deviation), mean + 3(standard deviation))

**Income Variable**

```python
mean = np.mean(superstore_df.Income)
median = np.median(superstore_df.Income)
mode = st.mode(superstore_df.Income)
variance = np.var(superstore_df.Income)
standard_deviation = np.std(superstore_df.Income)

print("Mean is",mean)
print("Median is",median)
print("Mode is",mode)
print("Variance is",variance)
print("Standard deviation is",standard_deviation)
```

```python
lb = mean - 3*standard_deviation
ub = mean + 3*standard_deviation

# checking whether any value is an outlier
superstore_df[(superstore_df["Income"]<lb) | (superstore_df["Income"]>ub)]
```

No outliers in Income variable

# **Binomial Distribution**

Binomial probability distribution is the probability distribution showing the probability of given number of successes in a fixed number of independent trials .In binomial distribution first we will find success probabilities related to variables and create a formula. After inserting any values in the formula we may get our probability

**Kidhome Variable**

First we will find probability of success in each event. Here success is customers having a kids home and failure is not having one.

Success probability = total number of customers having atleast one kid home /
                      total number of customers

```python
# finding kid homes count of customers
superstore_df.Kidhome.value_counts()
```

```python
success_prob = (866+46)/(993+866+46)
round(success_prob,3)
```

We got our success probability . Now we will create a python code using stats function of scipy library to find our required probabilty

Formula = st.binom.pmf(k,n,p)

where

k = number of customers having kid home

n = total number of customers

p = probability of customer having atleast 1 kid home

here p = 0.479

Now we can put any value for k and n provided k<=n

Lets take an example

k = 5

n = 10

Here we will find what is the probability of 5 customers from a total of 10 customers of the company having atleast 1 kid home

```python
prob = st.binom.pmf(5,10,0.479)
prob
```

Here we found our probabilty as 0.243. Now we plot a line graph to show distribution of probabilities for different values of k and n = total number of customers.

```python
n = 1905
k_values = list(range(n+1))
p = 0.479
probabilities = [st.binom.pmf(k,n,p) for k in k_values]
sns.lineplot(x = k_values,y = probabilities)
plt.show()
```

The graph shows probability increasing sharply after 750 customers and decreasing sharply before 1000 customers

```python
expected_value = st.binom.mean(n,p)
expected_value
```

Our expected value is 912.495 for Kidhome variable

#**Normal Distribution**

Normal distribution is a continuous distribution which is symmetric around the mean of the numeric data i.e. it has zero skewness. We have considered some variables as following normal distributions. We will find their related probabilities and expected values. Expected value is nothing but mean of the distribution

**Income Variable**

Formula = st.norm.cdf(k,loc=mean,scale=std)

k = finding probability of getting any value less than this value

loc = mean of distribution

scale = standard deviation


```python
# finding mean and standard deviation
mean = np.mean(superstore_df.Income)
std = np.std(superstore_df.Income)

print(f"mean(expected value) is {mean}")
print(f"standard deviation is {std}")
```

Lets take an example of k = 40000

```python
# finding probability

k = 40000
prob = st.norm.cdf(k,loc=mean,scale=std)
print(f"Probability of getting any value less than 40000 is {prob}")
```

Plotting a graph for different values of k

```python
# maximum value of income variable
n = int(max(superstore_df.Income))
k_values = list(range(n+1))
probabilities = [st.norm.cdf(k,loc=mean,scale=std) for k in k_values]
sns.lineplot(x = k_values, y = probabilities)
plt.xlabel("Income")
plt.ylabel("Probabilities of less than income values")
plt.show()
```

Probabilities increasing sharply for income values 20000 - 80000 and after that normalising
