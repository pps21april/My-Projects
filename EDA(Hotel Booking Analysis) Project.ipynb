### Data Wrangling Code

1 - Handling Null Values


````python
# Write your code to make your dataset analysis ready.


# displays the number of null values in each column
hotel_booking_df.isnull().sum()
````

````python
# dropping unwanted columns

# dropping columns 'agent' and 'company' since they have high number of null values
hotel_booking_df.drop(columns=['agent','company'],inplace=True)

#diplaying column names of the dataset after removing columns
hotel_booking_df.columns
````

````python
# removing rows

# dropping rows having none values in 'country' and 'children' column
hotel_booking_df.dropna(subset = ['country','children'],inplace=True)

#diplaying dataset after removing rows
hotel_booking_df
````

````python
# checking whether there is any null value left
hotel_booking_df.isnull().values.any()
````

False indicates we have successfully removed all the null values . Now we will remove all the duplicate rows

2 - Removing Duplicate Rows

````python
# removing all the duplicate rows
hotel_booking_df.drop_duplicates(inplace=True)

#displaying number of rows after deleting duplicates
len(hotel_booking_df.index)
````

Number of rows successfully decreased from a number of 118898 to 86914 . Now we will handle the outliers.

3 - Handling Outliers

For handling outliers first we have to find information about elements of each column and find out whether there are any odd values . After that we will handle such odd values . To minimise space usage we will print the code of only such columns which have odd values

````python
#finding values which have frequency of 1 in lead_time
hotel_booking_df['lead_time'].value_counts()[hotel_booking_df['lead_time'].value_counts().values==1]
````

Here we find two values above 700 which seems odd . Now we have to find whether there are any other values greater than 700.


````python
hotel_booking_df[hotel_booking_df['lead_time']>700]
````

Since there are no other values greater than 700 we will remove rows having values greater than 700 in 'lead_time' column

````python
#removing rows with lead time greater than 700
hotel_booking_df.drop(hotel_booking_df[hotel_booking_df['lead_time']>700].index,inplace=True)

# printing number of rows after removing outliers
len(hotel_booking_df.index)
````

````python
#finding outliers in reserved room type
print(hotel_booking_df['reserved_room_type'].value_counts())

#finding outliers in assigned room type
print(hotel_booking_df['assigned_room_type'].value_counts())
````

Here we find L and P types as outliers in both . We will remove them from dataset

````python
#removing rows with reserved room type L or P
hotel_booking_df.drop(hotel_booking_df[(hotel_booking_df['reserved_room_type']=='L') | (hotel_booking_df['reserved_room_type']=='P')].index,inplace=True)

#removing rows with assigned room type L or P
hotel_booking_df.drop(hotel_booking_df[(hotel_booking_df['assigned_room_type']=='L') | (hotel_booking_df['assigned_room_type']=='P')].index,inplace=True)


#printing number of rows after removal
len(hotel_booking_df.index)
````

````python
#finding outliers in distribution channel
hotel_booking_df['distribution_channel'].value_counts()
````

Undefined is an outlier here. We have to remove it

````python
#removing row with undefined value in distribution channel
hotel_booking_df.drop(hotel_booking_df[hotel_booking_df['distribution_channel']=='Undefined'].index,inplace=True)
````

````python
#removing row with 5400 value in adr column
hotel_booking_df.drop(hotel_booking_df[hotel_booking_df['adr']==5400].index,inplace=True)
````

````python
#removing rows with Undefined element in mean column
hotel_booking_df.drop(hotel_booking_df[hotel_booking_df['meal']=="Undefined"].index,inplace=True)
````

````python
# Chart - 13 visualization code
#series showing total cancelled bookings for each number of children
cancelled = hotel_booking_df.groupby("children")["is_canceled"].sum()
#sorting the index of series in alphabetical order
cancelled.sort_index(inplace=True)

#series showing total bookings for each number of children
counts = hotel_booking_df["children"].value_counts()
#sorting the index of series in alphabetical order
counts.sort_index(inplace=True)


#plotting percent of total bookings cancelled against each number of children
#percent is calculated by dividing values of cancelled bookings with total bookings multiplied by 100
sns.barplot(x = cancelled.index.astype(int) , y = (cancelled.values/counts.values)*100)
plt.xlabel("Number of Children")
plt.ylabel("% Booking Cancelled")
plt.title("% Bookings Cancelled for Each Number of Children")
plt.show()
````
